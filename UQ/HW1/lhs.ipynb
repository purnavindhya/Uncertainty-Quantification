```python
import numpy as np
import matplotlib.pyplot as plt
```

1. As we move across the evenly spaced grid with N spacings and N+1 edges, in each sub-grid, a sample has to be chosen.
2. For each sample, n, a random vector of dimension 'd' has to be sampled from the uniform distribution with limits [n/N, (n+1)/N] 
3. Once all the points are sampled, we've to make sure that only one point is sampled from each of [n/N, (n+1)/N]. Hence, shuffle all the points in each dimension.

```python
def latinhydercubesampling(N,d):   
    random_points = np.zeros(shape= (N,d))
    for i in range(N):
        low_ = i/N # lower limit of the i-th interval
        high_ = (i+1)/N # upper limit of the i-th interval
        pt = np.random.uniform(low= low_, high= high_, size=[1,d]) # generating a sample in d-dimension
        random_points[i,:] = pt # storing the set of samples in a random_points
    
    for dim in range(d):
        np.random.shuffle( random_points[:,dim] ) # shuffling the points so that there are no two points sampled from the same interval

    return random_points

```

```python
N = 10
d = 2
samples = latinhydercubesampling(N,d)
plt.figure(figsize=[10,10])
plt.xlim([0,1])
plt.ylim([0,1])
plt.scatter(samples[:,0],samples[:,1],c='r')

for i in np.arange(0,1,1/N):
    plt.axvline(i)
    plt.axhline(i)
plt.show
```
